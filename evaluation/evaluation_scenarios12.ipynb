```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

sns.set(color_codes=True, font_scale=1.4, style="whitegrid")

df = pd.read_csv("scenario1-small/2018-09-18-09-29-29/nginx-wrk")
df = df.set_index('run')

df2 = pd.read_csv("scenario1-small/2018-09-18-15-31-31/nginx-wrk")
df2 = df2.set_index('run')

df3 = pd.read_csv("scenario2-big/2018-09-18-09-29-12//nginx-wrk")
df3 = df3.set_index('run')

df4 = pd.read_csv("scenario2-big/2018-09-18-15-31-50/nginx-wrk")
df4 = df4.set_index('run')

frames = [df, df2]
frames2 = [df3, df4]

result = pd.concat(frames, ignore_index=True)
result2 = pd.concat(frames2, ignore_index=True)

y_values = ["wrk-json-requests", "wrk-json-bytes", "wrk-json-latency-99percentile"]
y_names = ["Requests (#)", "Traffic (Bytes)", "99 Percentile of Latency"]

plots = []
```

```python
for index, y_value in enumerate(y_values):
    plots.append(sns.catplot(x="vcpus:nginx-vnf", y=y_value, kind="bar", hue="cpu_time:nginx-vnf", data=result));
    plots[len(plots)-1].set(xlabel='vCPU # - NGINX', ylabel=y_names[index])
    plots[len(plots)-1]._legend.set_title("CPU Time - NGINX")
    
for y_value in y_values:
    plots.append(sns.catplot(x="memory:nginx-vnf", y=y_value, kind="bar", data=result));
    plots[len(plots)-1].set(xlabel='Memory (MB) - NGINX', ylabel=y_names[index])
    
```

```python
for index, y_value in enumerate(y_values):
    plots.append(sns.catplot(x="vcpus:nginx-vnf", y=y_value, kind="bar", hue="cpu_time:nginx-vnf", data=result2));
    plots[len(plots)-1].set(xlabel='vCPU # - NGINX', ylabel=y_names[index])
    plots[len(plots)-1]._legend.set_title("CPU Time - NGINX")
    
for y_value in y_values:
    plots.append(sns.catplot(x="memory:nginx-vnf", y=y_value, kind="bar", data=result2));
    plots[len(plots)-1].set(xlabel='Memory (MB) - NGINX', ylabel=y_names[index])
```
