```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

plt.figure(figsize=(9, 9))

sns.set(color_codes=True, font_scale=1.4, style="whitegrid")

df = pd.read_csv("scenario3-small/haproxy-nginx-wrk")
df = df.set_index('run')

df2 = pd.read_csv("scenario4-big/haproxy-nginx-wrk")
df2 = df2.set_index('run')

y_values = ["wrk-json-requests", "wrk-json-bytes", "wrk-json-latency-99percentile"]
y_names = ["Requests (#)", "Traffic (Bytes)", "99 Percentile of Latency"]

plots = []
```

```python
for index, y_value in enumerate(y_values):
    plots.append(sns.catplot(x="vcpus:nginx-vnf", y=y_value, kind="bar", hue="cpu_time:nginx-vnf", data=df));
    plots[len(plots)-1].set(xlabel='vCPU # - NGINX', ylabel=y_names[index])
    plots[len(plots)-1]._legend.set_title("CPU Time - NGINX")
    plots.append(sns.catplot(x="vcpus:haproxy-vnf", y=y_value, kind="bar", hue="cpu_time:haproxy-vnf", data=df));
    plots[len(plots)-1].set(xlabel='vCPU # - HAProxy', ylabel=y_names[index])
    plots[len(plots)-1]._legend.set_title("CPU Time - HAProxy")
    
for y_value in y_values:
    plots.append(sns.catplot(x="memory:nginx-vnf", y=y_value, kind="bar", data=df));
    plots[len(plots)-1].set(xlabel='Memory (MB) - NGINX', ylabel=y_names[index])
    plots.append(sns.catplot(x="memory:haproxy-vnf", y=y_value, kind="bar", data=df));
    plots[len(plots)-1].set(xlabel='Memory (MB) - HAProxy', ylabel=y_names[index])
    
```

```python
for index, y_value in enumerate(y_values):
    plots.append(sns.catplot(x="vcpus:nginx-vnf", y=y_value, kind="bar", hue="cpu_time:nginx-vnf", data=df2));
    plots[len(plots)-1].set(xlabel='vCPU # - NGINX', ylabel=y_names[index])
    plots[len(plots)-1]._legend.set_title("CPU Time - NGINX")
    plots.append(sns.catplot(x="vcpus:haproxy-vnf", y=y_value, kind="bar", hue="cpu_time:haproxy-vnf", data=df2));
    plots[len(plots)-1].set(xlabel='vCPU # - HAProxy', ylabel=y_names[index])
    plots[len(plots)-1]._legend.set_title("CPU Time - HAProxy")

for y_value in y_values:
    plots.append(sns.catplot(x="memory:nginx-vnf", y=y_value, kind="bar", data=df2));
    plots[len(plots)-1].set(xlabel='Memory (MB) - NGINX', ylabel=y_names[index])
    plots.append(sns.catplot(x="memory:haproxy-vnf", y=y_value, kind="bar", data=df2));
    plots[len(plots)-1].set(xlabel='Memory (MB) - HAProxy', ylabel=y_names[index])
    
```
